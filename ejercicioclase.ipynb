```python
import pandas as pd

context = pd.read_csv(r"C:\Users\ferna\Desktop\Apuntes Clase\Ejercicios python caseros\fitlife_context.csv")
members = pd.read_csv(r"C:\Users\ferna\Desktop\Apuntes Clase\Ejercicios python caseros\fitlife_members.csv")

print(context.head())
print(members.head())
```

```python
df = pd.merge(context, members, on="month", how="left")

df.shape
df.columns
df.head()
```

```python
df.info()
df.describe()
```

```python
# Distribución de los clientes por plan
df["plan"].value_counts(normalize=True)
# De los 3 planes parece que el basic no es el principal
```

```python
# Ingresos por plan
df.groupby("plan")["price_paid"].sum()
# El basic y el family parecen los que mas ingresos generan para el negocio. 
```

```python
# Margen por plan 
df["margin"] = df["price_paid"] - df["cost_to_serve"]
df.groupby("plan")["margin"].mean()
# Esto es ORO analítico. Significa: Basic = menor margen, Family = producto estrella en rentabilidad y Premium = equilibrio volumen + margen
```
